```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('20100101', periods=1000))
ts.head()
```

```python
ts = ts.cumsum()
ts.head()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('abcd'))
df = df.cumsum()
```

```python
plt.figure(figsize=(10, 6))
df.plot()
```

Series 只有一列，所以plot() 产生一条线；
DataFrame 有多列，所以plot() 产生多条线；

```python
df.plot(y='b')
```

```python
ser = pd.Series(list(range(1000)))
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['b', 'c']).cumsum()
df3['a'] = ser
df3.head()
```

```python
df3.plot(x='a', y='b')   #x 需要Series 类型，自己定义一个list 是不行的，如下
```

```python
df3.plot(x=list(range(1000)), y='b')
```

The **by** keyword can be specified to plot grouped histograms:

```python
import pandas as pd
import numpy as np
data = pd.Series(np.random.randn(1000))
```

```python
data.hist(by=np.random.randint(0, 4, 1000))   #把data里的每个数据标记一个tag，tag为0,1,2,3，按照tag进行分组绘图
```

### Scatter 图

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=list("ABCD"))
df
```

```python
df.plot.scatter(x='A', y='B')
```

需要展示多列，可使用ax 定义多个，如下

```python
ax = df.plot.scatter(x='A', y='B', color='r', label='Group 1')
df.plot.scatter(x='C', y='D', color='k', label='Group 2', ax=ax)
```

```python
df.plot.scatter(x='A', y='B', c='C', s=500);
```

```python
df.plot.scatter(x='A', y='B',s=df['C']*200);
```

### pie 饼图

```python
series = pd.Series(3 * np.random.rand(4), index=list('abcd'), name='series')
series.plot.pie(figsize=(6,6))
```

df类型的数据画饼图有 2 种方法，1. 用 y 参数；2. 设定 subplots=True

```python
df = pd.DataFrame(3 * np.random.rand(4,2), index=list("abcd"), columns=['X', 'Y'])
df.plot.pie(subplots=True, figsize=(8, 4))
```

```python
df.plot.pie(y='X', figsize=(4,4), legend=False)  #隐藏图例
```

```python
df.plot.pie(y='Y', labels=None)  #去掉标签
```

```python
series.plot.pie(labels=['AA', 'BB', 'CC', 'DD'], colors=['r', 'g', 'b', 'c'],
   ....:                 autopct='%.2f', fontsize=20, figsize=(6, 6))
```

如果sum total 小于 10，会变成扇形图

```python
series = pd.Series([0.1]*4, index=list('abcd'), name='haha')
series.plot.pie(figsize=(6,6))
```
